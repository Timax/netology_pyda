**Задание 2** <br>
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
df2 = pd.read_csv('power.csv')
#df2.head(20)
#df2.country.unique()
```

```python
baltics = ['Estonia','Latvia','Lithuania']
categories = [4,12,21]
```

```python
df2[
    (df2['country'].isin(baltics))
    & (df2['category'].isin(categories))
    & (df2['year'].between(2005,2010))
    & (df2['quantity'] >= 0)
]['quantity'].sum()
#df2[(df2['country'].str.contains('|'.join(baltics))),
#df2[(df2['country'] == 'Estonia') | (df2['country'] == 'Latvia') | (df2['country'] == 'Lithuania')]
```
