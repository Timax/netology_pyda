***Задание 1***\
Дана статистика рекламных каналов по объемам продаж.\
Напишите скрипт, который возвращает название канала с максимальным объемом.\
Т. е. в данном примере скрипт должен возвращать “yandex”:\
stats = {“facebook”: 55, “yandex”: 120, “vk”: 115, “google”: 99, “email”: 42, “ok”: 98}

```python
stats = {'facebook': 55, 'yandex': 120, 'vk': 115, 'google': 99, 'email': 42, 'ok': 98}
max_volume_value = 0
result = ''
for company in stats.keys():
    if stats[company] > max_volume_value:
        max_volume_value = stats[company]
        result = company
print(result)
```

***Задание 3***\
В примере поиска с помощью кортежей мы использовали 3 столбца. Напишите функцию, которая формирует словарь для поиска по n столбцам.

```python
# функция созадания словаря из файла stats.csv по заданному числу элементов n
def dictionary(n):
    stats_dict = {}    
    with open ('stats.csv') as f:
        for line_in_file in f:
            elem = line_in_file.strip().split(',')             
            stats_dict.setdefault(tuple(elem[:n]),0)
            #присваиваем некое значение:выбираю все, что осталось в строке после определения ключа
            stats_dict[tuple(elem[:n])] = elem[n:]
    return stats_dict
#вызов функции с указанным значением числа элементов n и отображение первых 100 элементов словаря для проверки
#перезаписи по ключу
n = 2
i = 0
for key, value in dictionary(n).items():
    print(key, ':',value)
    i +=1
    if i == 100:
        break
        
#s = sorted(dictionary(2).items(), key = lambda k: k[0])[:5]
#вывод 5 наименьших ключей
#print(type(s))

```
